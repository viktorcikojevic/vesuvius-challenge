```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load training_log.csv to a pandas dataframe
df = pd.read_csv('metrics.csv')
df
```

```python
# Plot loss and val_loss vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['step'], df['loss'], label='loss')
plt.plot(df['step'], df['val_loss'], label='val_loss')
plt.legend()
```

```python
# plot accuracy and val_accuracy vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['step'], df['accuracy'], label='accuracy')
plt.plot(df['step'], df['val_accuracy'], label='val_accuracy')

plt.ylim(0.6, 1.1)
# set horizontal line at 1.0
plt.axhline(1.0, color='red', linestyle='--')

plt.legend()
```

```python
# plot precision and val_precision vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['step'], df['precision'], label='precision')
plt.plot(df['step'], df['val_precision'], label='val_precision')

plt.ylim(0, 1.1)
# set horizontal line at 1.0
plt.axhline(1.0, color='red', linestyle='--')

plt.legend()
```

```python
# plot f0point5 and val_f0point5 vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['step'], df['f0point5'], label='f0point5')
plt.plot(df['step'], df['val_f0point5'], label='val_f0point5')

plt.ylim(0, 1.1)

# set horizontal line at 1.0
plt.axhline(1.0, color='red', linestyle='--')

plt.legend()
```
