```python
import numpy as np
```

```python
# list files in train/ dir
files = !ls train/volume
files[:10]
```

```python
import matplotlib.pyplot as plt
x = np.load(f"train/volume/{files[40]}")

# visualize the first image
plt.imshow(x[:,:,0], cmap='gray')
```

```python
avgs = []
stds = []
for i in range(100):
    x = np.load(f"train/label/{files[i]}")
    avg, std = x.mean(), x.std()
    avgs.append(avg)
    stds.append(std)
```

```python
import matplotlib.pyplot as plt

plt.hist(avgs, bins=100);
```

```python
plt.hist(stds, bins=100);
```
